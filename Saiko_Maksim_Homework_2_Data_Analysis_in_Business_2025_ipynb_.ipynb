##Assignment for Part 2:

4. Use a pre-trained BERT model to classify the sentiment of movie reviews.(Task weight: 20%)

```python
print("Loading and Preparing IMDB Dataset...")
print(f"Initial dataset shape: {imdb_df.shape}")

print(f"Missing values: {imdb_df.isnull().sum().sum()}")

sentiment_counts = imdb_df['sentiment'].value_counts()
print(f"Sentiment distribution:\n{sentiment_counts}")

balance_ratio = sentiment_counts['positive'] / sentiment_counts['negative']
print(f"Positive/Negative ratio: {balance_ratio:.3f}")

def preprocess_for_bert(text):
    text = str(text)
    text = re.sub(r'<br\s*/?>', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

imdb_df['cleaned_review'] = imdb_df['review'].apply(preprocess_for_bert)

imdb_df['label'] = imdb_df['sentiment'].map({'positive': 1, 'negative': 0})

sample_size = 2000
if len(imdb_df) > sample_size:
    df_sample = imdb_df.sample(n=sample_size, random_state=42)
else:
    df_sample = imdb_df

print(f"\nUsing sample of {len(df_sample)} reviews for training")

train_texts, test_texts, train_labels, test_labels = train_test_split(
    df_sample['cleaned_review'].tolist(),
    df_sample['label'].tolist(),
    test_size=0.2,
    random_state=42,
    stratify=df_sample['label']
)

print(f"Training set: {len(train_texts)} reviews")
print(f"Test set: {len(test_texts)} reviews")

MODEL_NAME = 'bert-base-uncased'

try:
    tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)
    model = BertForSequenceClassification.from_pretrained(
        MODEL_NAME,
        num_labels=2,
        output_attentions=False,
        output_hidden_states=False
    )
    print(f"Successfully loaded {MODEL_NAME}")
    print(f"Model architecture: {model.__class__.__name__}")
    print(f"Number of parameters: {sum(p.numel() for p in model.parameters()):,}")

except Exception as e:
    print(f"Error loading model: {e}")
    print("Trying alternative approach...")
    MODEL_NAME = 'distilbert-base-uncased'
    tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)
    model = BertForSequenceClassification.from_pretrained(
        MODEL_NAME,
        num_labels=2
    )
    print(f"Loaded fallback model: {MODEL_NAME}")

train_encodings = tokenizer(
    train_texts,
    truncation=True,
    padding=True,
    max_length=128,
    return_tensors='pt'
)

test_encodings = tokenizer(
    test_texts,
    truncation=True,
    padding=True,
    max_length=128,
    return_tensors='pt'
)

print(f"Tokenization completed")
print(f"Input IDs shape - Train: {train_encodings['input_ids'].shape}")
print(f"Attention mask shape - Train: {train_encodings['attention_mask'].shape}")

class IMDBDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {
            key: val[idx] for key, val in self.encodings.items()
        }
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = IMDBDataset(train_encodings, train_labels)
test_dataset = IMDBDataset(test_encodings, test_labels)

print(f"\nDataset created:")
print(f"  Train dataset size: {len(train_dataset)}")
print(f"  Test dataset size: {len(test_dataset)}")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
if device.type == 'cuda':
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024**2:.1f} MB")

model.to(device)

import transformers
print(f"Transformers version: {transformers.__version__}")

if hasattr(TrainingArguments, 'evaluation_strategy'):
    training_args = TrainingArguments(
        output_dir='./results',
        num_train_epochs=3,
        per_device_train_batch_size=8,
        per_device_eval_batch_size=8,
        warmup_steps=100,
        weight_decay=0.01,
        logging_dir='./logs',
        logging_steps=10,
        evaluation_strategy='epoch',
        save_strategy='epoch',
        load_best_model_at_end=True,
        metric_for_best_model='f1',
        greater_is_better=True,
    )
else:
    training_args = TrainingArguments(
        output_dir='./results',
        num_train_epochs=3,
        per_device_train_batch_size=8,
        per_device_eval_batch_size=8,
        warmup_steps=100,
        weight_decay=0.01,
        logging_dir='./logs',
        logging_steps=10,
        eval_strategy='epoch',
        save_strategy='epoch',
        load_best_model_at_end=True,
        metric_for_best_model='f1',
        greater_is_better=True,
    )

def compute_metrics(p):
    preds = np.argmax(p.predictions, axis=1)
    precision, recall, f1, _ = precision_recall_fscore_support(
        p.label_ids, preds, average='binary'
    )
    acc = accuracy_score(p.label_ids, preds)

    return {
        'accuracy': acc,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics,
)

trainer.train()

predictions = trainer.predict(test_dataset)
preds = np.argmax(predictions.predictions, axis=1)

accuracy = accuracy_score(test_labels, preds)
precision, recall, f1, _ = precision_recall_fscore_support(
    test_labels, preds, average='binary'
)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")
print("\nDetailed Classification Report:")
print(classification_report(test_labels, preds,
                           target_names=['Negative', 'Positive']))

def predict_sentiment(text, model, tokenizer, device):
    text = preprocess_for_bert(text)
    inputs = tokenizer(
        text,
        truncation=True,
        padding=True,
        max_length=128,
        return_tensors='pt'
    ).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        pred = torch.argmax(probs, dim=-1).item()
        confidence = probs[0][pred].item()

    sentiment = 'positive' if pred == 1 else 'negative'
    return sentiment, confidence

test_examples = ["Great.", "Bad.", "Ok.", "Phenomenal.", "Horrible."]

for i, review in enumerate(test_examples, 1):
    sentiment, confidence = predict_sentiment(review, model, tokenizer, device)
    print(f"\nExample {i}:")
    print(f"Review: {review[:100]}...")
    print(f"Prediction: {sentiment.upper()} (confidence: {confidence:.2%})")

metrics = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Value': [accuracy, precision, recall, f1]
}
metrics_df = pd.DataFrame(metrics)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

colors = ['#2ecc71', '#3498db', '#e74c3c', '#f39c12']
bars = axes[0].bar(metrics_df['Metric'], metrics_df['Value'], color=colors)
axes[0].set_ylim([0, 1])
axes[0].set_ylabel('Score')
axes[0].set_title('BERT Model Performance Metrics', fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)

for bar in bars:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2., height + 0.02,
                f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_labels, preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negative', 'Positive'],
            yticklabels=['Negative', 'Positive'], ax=axes[1])
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')
axes[1].set_title('Confusion Matrix', fontweight='bold')

plt.tight_layout()
plt.show()

try:
    if hasattr(trainer.state, 'log_history'):
        log_history = trainer.state.log_history

        train_loss = [x['loss'] for x in log_history if 'loss' in x]
        eval_loss = [x['eval_loss'] for x in log_history if 'eval_loss' in x]

        if train_loss:
            fig, axes = plt.subplots(1, 2, figsize=(12, 4))

            axes[0].plot(train_loss, 'b-', linewidth=2)
            axes[0].set_xlabel('Training Steps')
            axes[0].set_ylabel('Loss')
            axes[0].set_title('Training Loss', fontweight='bold')
            axes[0].grid(True, alpha=0.3)

            eval_accuracy = [x.get('eval_accuracy', 0) for x in log_history if 'eval_accuracy' in x]
            if eval_accuracy:
                axes[1].plot(eval_accuracy, 'r-', linewidth=2)
                axes[1].set_xlabel('Evaluation Steps')
                axes[1].set_ylabel('Accuracy')
                axes[1].set_title('Evaluation Accuracy', fontweight='bold')
                axes[1].grid(True, alpha=0.3)
            else:
                axes[1].text(0.5, 0.5, 'No evaluation metrics\navailable in history',
                           ha='center', va='center', transform=axes[1].transAxes)
                axes[1].set_title('Evaluation Metrics', fontweight='bold')

            plt.tight_layout()
            plt.show()
except Exception as e:
    print(f"Could not plot training history: {e}")

model_save_path = './sentiment_model'
trainer.save_model(model_save_path)
tokenizer.save_pretrained(model_save_path)
print(f"Model saved to {model_save_path}")
```
